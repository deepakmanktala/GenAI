STACK Implementations


```python
## Python list provides methods such as append and pop whihc can be used as Stack methods

my_stack = []

my_stack.append('a')
my_stack.append('b')
my_stack.append(1)

print(my_stack)

my_stack.pop()
print(my_stack)

print('\nElements popped from stack:')
print(my_stack.pop())
print(my_stack.pop())
print(my_stack.pop())
```

```python
## Using Collections
from collections import deque

stack = deque()

stack.append('a')
stack.append('b')
stack.append('c')

print('Initial Stack :')
print(stack)

print('\nElements popped from stack:')
print(stack.pop())
print(stack)
print(stack.pop())
print(stack)
print(stack.pop())

print('\nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())
```
