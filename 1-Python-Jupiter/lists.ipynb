LIST COMPREHENSIONS

```python
### list comprehensions

lst = []
for x in range(10):
    lst.append(x**2)

print(lst)



```

```python
n100 = []

for i in range(1, 101):
    n100.append(i)

for num in n100:
    if num < 2:
        print(num, "is not a prime")
        continue

    for j in range(2, num):
        if num % j == 0:
            print(num, "is not a prime")
            break
    else:
        print(num, "is a prime")
```

LIST COMPREHENSIONS

basic syntax [expression for the item in iterable ]

with conditional logic [ expression for item  in iterable with condition]

Nested List Comprehensions --> iterate through two different iterables

```python
## LIST COMPREHENSIONS

square = [num**2 for num in range(10)]
print(square)

## List comprehension with condition
lst = []
for i in range(10):
    if i%2==0:
        lst.append(i)

```

```python
even_number = [num for num in range(10) if num%2==0]
print(even_number)
```

```python
## Nested List comprehension
list1 = [1,2,3,4,5,6,78,8]
list2 = ['a','b','b','c','d','e','f','g','h','i']

pair = [(i,j) for i in list1 for j in list2]

print(pair)
```

```python
for num in range(1, 101):
    if num > 1:
        is_prime = True

        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            print(num)
```

```python
for num in range(2, 11):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num)


for num in range(2, 11):
    is_prime = True

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```

```python
def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, num):
        if num % i == 0:
            return False

    return True


for num in range(1, 11):
    if is_prime(num):
        print(num)


```

```python


primes = [
    num
    for num in range(2, 11)
    if all(num % i != 0 for i in range(2, num))
]

print(primes)
```

```python
def is_prime(num):
    if num <= 1:
        return False

    for i in range(2, num):
        if num % i == 0:
            return False

    return True


primes = list(filter(is_prime, range(1, 11)))

print(primes)
```

```python
n = 10

prime = [True] * (n + 1)

prime[0] = False
prime[1] = False

for i in range(2, int(n ** 0.5) + 1):
    if prime[i]:
        for multiple in range(i * i, n + 1, i):
            prime[multiple] = False

for i in range(1, n + 1):
    if prime[i]:
        print(i)
```

```python
def is_prime(num, divisor=2):
    if num <= 1:
        return False

    if divisor == num:
        return True

    if num % divisor == 0:
        return False

    return is_prime(num, divisor + 1)


for num in range(1, 11):
    if is_prime(num):
        print(num)
```
