RealWordlExamples Using LIST in PYTHON

```python
to_do_list = ["Buy groceries", "Work on finding a job", "Clean the house", "take showwer", "have food", "study for the job"]

## adding another tasks to the list

to_do_list.append("interview ready")
to_do_list.append("Go for a walk")
to_do_list.append("Play with Alizeh")


## Remove the completed task

to_do_list.remove("take showwer")

## Checking if task is in the list

if "Buy groceries" in to_do_list:
    print("Don't forget the groceries")

print("To DO List remaining")

for task in to_do_list:
    print(f"- {task}")
```

ORGANIZING STIDENT GRADES

```python
grades = [80, 76,89,98,65, 76, 79]

grades.append(95)

average_grade = sum(grades)/len(grades)

print(average_grade)

print(f"Average Grade : {average_grade:.2f}")

highest_grade = max(grades)
lowest_grade = min(grades)

print(f"Highest Grade : {highest_grade}, Lowest Grade: {lowest_grade}")
```

```python
# Managing an inventory

# Initial inventory
inventory = ["apples", "bananas", "oranges", "grapes"]

print("Initial inventory:", inventory)

# Adding a new item
inventory.append("strawberries")
print("After adding strawberries:", inventory)

# Removing an item that is out of stock
inventory.remove("bananas")
print("After removing bananas:", inventory)

# Checking if an item is in stock
item = "oranges"

if item in inventory:
    print(f"{item} are in stock.")
else:
    print(f"{item} are out of stock.")

# Display final inventory
print("Final inventory:", inventory)
print("Inventory List")
for item in inventory:
    print(f"- {item}")
```

```python
# Collecting user feedback
feedback = [
    "Great service!",
    "Very satisfied",
    "Could be better",
    "Excellent experience"
]

# Adding new feedback
feedback.append("Not happy with the service")

# Counting specific feedback (positive keywords)
positive_feedback_count = sum(
    1 for comment in feedback
    if "great" in comment.lower() or "excellent" in comment.lower()
)

print(f"Positive Feedback Count: {positive_feedback_count}")

# Printing all feedback
print("\nUser Feedback:")
for comment in feedback:
    print(f"- {comment}")
```
