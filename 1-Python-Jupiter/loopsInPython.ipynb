```python
# Loop Contrl Statements

## break
## the break statement exits the loop prematurely

for i in range(10):
    if i == 5:
        break
    print(i)



```

```python
## continue statement , where to continue
## continue statement skips the current iteration and continue with the next

## print odd chars
for i in range(12):
    if i%2 == 0:
        continue
    print(i)
    # else:
    #     print(i)


## print even chars
for i in range(1,9):
    if i%2 != 0:
        continue
    print(i)


## PASS --> pas is just to pass , it doe snothing

```

NESTED LOOPS

```python
## NESTED LOOPS
## A loop inside a loop

for i in range(5):
    for j in range(4):
        # print("i:{i} and j:{j}", i, j)
        print(f"(i is {i} and j is {j})")
```

```python
## Eamplese
## Calculate the sum of first n natural numbers with while and for loops independently

n = 200

sum = 0
while n > 0:
    sum = sum + n
    n = n - 1
    print("this is in between sum {sum}and the value of n is {n}", sum,n)
print(sum)

```

```python
### Now lets calculate the sum with for statement in a loop
n = 20
sum = 0 ## temporary variable
# count = 1 ## This would be a redundant step, not needed
for n in range(1,n+1):
    sum = sum + n
print(sum)
```

```python
## Example : Display the prime numbers b/w 1 an 100
## Do it using both while and for loops

n = 100

# for i in range(2,n+1):
#     for j in range(2,i):
#         print(f"I came here and i is {i}, j is {j}")

for i in range(2, n + 1):
    is_prime = True
    
    for j in range(2, i):
        if i % j == 0:
            is_prime = False
            break
    
    if is_prime:
        print(i, "is a prime number")
    else:
        print(i, "is not a prime number")




```

```python
n = 100

for i in range(2, n + 1):
    is_prime = True
    
    for j in range(2, int(i**0.5) + 1):
        if i % j == 0:
            is_prime = False
            break
    
    if is_prime:
        print(i)
```
