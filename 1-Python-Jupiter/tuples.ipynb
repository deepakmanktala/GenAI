```python
## Tuple Operations - Concatenation, repetition, slicing, 
numb_tup = (1,2,3,4,5,6,7,8,9,9)
mix_tup = ('a','b','c',1,2,3,4,5,6,7,1.5,3.5,True)

concat_tup = numb_tup +mix_tup
print(concat_tup)

con_tup_2 = mix_tup+numb_tup
print(con_tup_2)
```

```python
## Repetition
triple_tup = mix_tup * 5
print(triple_tup)

double_list = lst1*2
print(double_list)
```

```python
## Immutable nature of Tuples
## elements of a Tuple can not be changed once assigned
lst = [1,2,3,4,5,6,7,8]
print(lst)

lst[1] = "Deepak"
print(lst)

tup = (1,2,3,4,4,5,6,7,7,7)
print(tup)
tup[1] = "dedass"
print(tup)
```

```python
## packing and unpacking tuples

## packing a tuple
packed_tup = 1,"Hello", "Alizeh", True, "Pi", 3.14
print(packed_tup)

## unpacking a tuple

a,b,c,d,e,f = packed_tup

print(a,b,c,d,e,f)

for i in range (len(packed_tup)):
    print(packed_tup[i], end = " ")
    print(" ".join(map(str, packed_tup)))

    
```

```python
## Iterating over nested tuples

## nested tuples
nest_tup = ((1,2,3), ('a','b','c'), (1,2,True,'c','f'))
nest_list = [['a','c','v','d',1,'23a','sa'], [1,2,3,4,4], (12,3,323,32,343,43)]

print(nest_list[0][0:3])
print(nest_tup[2][::-1])

nest_tup_2 = ((1,2,3,32),['a', 'b',1,2,3,3],('a',1,32,32,'32sc','assa'),['33','dsds',323, 'sas'])
print(nest_tup_2)

print(nest_tup_2[2][2:4])

nest_tup_3 = ((1,2,32,'k'),['a','b','g','h'],(3,4,5,6,7,8,9,2120120))
print(nest_tup_3[2][-1])

for sub_tuple in nest_tup_3:
    for item in sub_tuple:
        print(item, end = " ")
    print()
```
