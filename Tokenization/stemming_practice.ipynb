```python
## Stemming --> Eating, eaten, eat, eatable --> stem word is eat
## program, programming, programmed --> stem word is program

words = ["eating", "eaten", "eat","finally", "finalized", "health", "healthy", "healthian", "healthier", "work", "working", "worked", "workable", "eats","history", "historic", "comment", "commenting", "commented", "excited", "exciting", "excitation"]
```

```python
## porter stemming
## porter stemmer
from nltk.stem import PorterStemmer
words = ["eating", "eaten", "eat","finally", "finalized", "health", "healthy", "healthian", "healthier", "work", "working", "worked", "workable", "eats","history", "historic", "comment", "commenting", "commented", "excited", "exciting", "excitation"]
stemming_1 = PorterStemmer()

for word in words:
    print(word + "-------->>"+ stemming_1.stem(word))
```

```python
stemming_1.stem("congratulations")
```

```python
stemming_1.stem("sitting")
```

Regexp Stemmer Class

```python
## Regexp stemmer Examples
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)

reg_stemmer.stem('eating')
reg_stemmer.stem('heated')

reg_stemmer.stem('ingeating')
```
