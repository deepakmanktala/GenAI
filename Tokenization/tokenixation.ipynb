```python
## Tokenization
## Pragraphs --> Sentences
from nltk.tokenize import sent_tokenize

corpus = """ Hello Welcome, welcome to my world. This is my NLP program.
Do complete it for brushing up the learnt and implemented items. Excited!
"""
print(corpus)

documents_1 = sent_tokenize(corpus)
print(documents_1)
```

```python
print(type(documents_1))
for sentence in documents_1:
    print(sentence)
```

```python
## Tokenization
## Paragraph --> Words
## Sentences --> Words
from nltk.tokenize import word_tokenize

word_tokens_1 = word_tokenize(corpus)
print(word_tokens_1)
```

```python
for sentence in documents_1:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tree_bank_tokens = TreebankWordTokenizer()
tree_bank_tokens.tokenize(corpus)
```
