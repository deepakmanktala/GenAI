```python
## Why do we need functions
num = 24
if num %2 == 0:
    print("The number is even")
else:
    print("The number is odd")
```

```python
def even_or_odd(num):
    if num %2 == 0:
        print("The number is even")
    else:
        print("The number is odd")

even_or_odd(24)
even_or_odd(45)
```

```python
## Default Parameters
def greet(name= "Guest" ):
    print(f"Hello {name}, Welcome to the party")

greet('Deepak')
greet()
greet('Alizeh')
```

```python
## Variable length arguments
## Positional and Keyword Arguments



## Positional Arguments Example
def print_numbers(*args):
    for number in args:
        print(number, end = " ")

print_numbers(1,2,3,3,4,4,4,4,43,3,3,333,3,33,33,323,333, "Deepak")
```

```python
## Keyword Argument --> All the params will be in key value pair , and it would be accessible similar to a dictionary
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}", end = " | ")

print_details(name= "Alizeh", age = 10, country = "US")

```

```python
## Combined positional arguments and keyword arguments

def print_details_combined(*args, **kwargs):

    for val in args:
        print(f" {val}", end = " ")

    print()

    for key, value in kwargs.items():
        print(f"{key}:{value}", end = " | ")

## Positional arguments shoudl always be before the keywprd arguments
print_details_combined(1,2,3,4,5,6,7,8,name = "Deepak", last_name = "manktala", age = 35, city = "BLR")
```

```python
## Return Statements
def multiply(a,b):
    return a*b

def multiply_2(a,b):
    return f"{a} * {b} = {a*b}"

print(multiply(5,7))
print(multiply_2(5,9))


def multiply(a, b):
    return a * b

def format_multiply(a, b):
    result = multiply(a, b)
    return f"{a} * {b} = {result}"

print(format_multiply(6,7))
```
