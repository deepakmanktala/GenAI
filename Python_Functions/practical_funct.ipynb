```python
from typing import Literal

TemperatureUnit = Literal["C", "F", "K"]


def convert_temperature(
    value: float,
    from_unit: TemperatureUnit,
    to_unit: TemperatureUnit,
    precision: int = 2
) -> str:
    """
    Convert temperature between Celsius, Fahrenheit, and Kelvin.

    Args:
        value (float): Temperature value to convert.
        from_unit (str): Source unit ("C", "F", "K").
        to_unit (str): Target unit ("C", "F", "K").
        precision (int): Decimal precision for output.

    Returns:
        str: Formatted conversion string.

    Raises:
        ValueError: If invalid unit is provided.
    """

    from_unit = from_unit.upper()
    to_unit = to_unit.upper()

    valid_units = {"C", "F", "K"}
    if from_unit not in valid_units or to_unit not in valid_units:
        raise ValueError(f"Invalid unit. Supported units: {valid_units}")

    # Step 1: Convert input to Celsius
    if from_unit == "C":
        celsius = value
    elif from_unit == "F":
        celsius = (value - 32) * 5 / 9
    elif from_unit == "K":
        celsius = value - 273.15

    # Step 2: Convert Celsius to target unit
    if to_unit == "C":
        result = celsius
    elif to_unit == "F":
        result = (celsius * 9 / 5) + 32
    elif to_unit == "K":
        result = celsius + 273.15

    # Step 3: Format output
    return f"{value:.{precision}f}°{from_unit} = {result:.{precision}f}°{to_unit}"
```

```python
## Password Strength Checker
def is_strong_password(password):
    """This fucntion checks if the password is strong enough"""
    if len(password) < 8:
        return False
    
    if not any(char.isdigit() for char in password):
        return False

    if not any(char.islower() for char in password ):
        return False

    if not any (char.isupper() for char in password):
        return False

    if not any (char in '!@#$%^&*()_+' for char in password):
        return False
    return True

print(is_strong_password("WekaPwd"))
print(is_strong_password("Strong@Password123"))
    
```

```python
## File operations
def count_word_frequency(file_path):
    word_count = {}
    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word= word.lower().strip('.,!?;:\'"')
                word_count[word]= word_count.get(word,0)+1
    return word_count

print(count_word_frequency('alizeh.txt'))

result = count_word_frequency("alizeh.txt")

# Top 5 most frequent words
top_5 = sorted(result.items(), key=lambda x: x[1], reverse=True)[:5]

for word, count in top_5:
    print(f"{word}: {count}")
```

```python
import re
from collections import Counter
from typing import Dict


def count_word_frequency(file_path: str) -> Dict[str, int]:
    """
    Count word frequency from a text file.

    Args:
        file_path (str): Path to the input text file.

    Returns:
        Dict[str, int]: Dictionary with word frequencies.

    Raises:
        FileNotFoundError: If file does not exist.
        IOError: If file cannot be read.
    """

    word_count = Counter()

    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            for line in file:
                # Extract words using regex (handles punctuation better)
                words = re.findall(r"\b\w+\b", line.lower())
                word_count.update(words)

    except FileNotFoundError:
        raise FileNotFoundError(f"File not found: {file_path}")
    except IOError as e:
        raise IOError(f"Error reading file: {e}")

    return dict(word_count)
```

```python
import re
## Validate the email address
def is_email_valid(email):
    """This function check if the email address is valid"""
    # pattern = r'^[a-zA-Z0-9_.+-]+@[a-zA-Z0-9-.]+$'
    pattern = r'^(?!.*\.\.)[A-Za-z0-9][A-Za-z0-9._%+-]*@[A-Za-z0-9-]+(\.[A-Za-z]{2,})+$'
    return re.match(pattern, email) is not None

print(is_email_valid('deepa'))
print(is_email_valid('deepa.manktala@gmail.com'))
print(is_email_valid('.com'))
print(is_email_valid('girdhar.a@gmail.com'))



```
