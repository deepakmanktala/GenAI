```python
## Mutiple Inheritance
## when a class inherits form more than one class


## bAse Class 1
class Animal:
    def __init__(self, name, category, life ):
        self.name = name
        self.category = category
        self.life = life
    
    def speak(self):
        print("Subclasses must implement this method")


## Base class 2
class Pet:
    def __init__(self, pet_name):
        self.pet_name = pet_name

    def giggle(self):
        print(f"This {self.pet_name} giigles")



## Derived Class:
class Cat(Animal, Pet):
    def __init__(self, name, category, life):
        Animal.__init__(self,name, category, life)
        Pet.__init__(self.pet_name)






    
```
