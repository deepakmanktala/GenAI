```python
## Instance variables and methods

class Cat:
    def __init__(self, name, age, talk):
        self.name= name
        self.age = age
        self.talk = talk
    

    
# Create CAT objects
north_cat = Cat("catty", 5 ,"Meaoun")
south_cat = Cat("billi", 7, 'Meow')  

print(north_cat.name)
print((south_cat.talk))
print(south_cat.name)
```

```python
## Class with instance methods

class Lion:
    def __init__(self, name, jungle, kids, age):
        self.name = name
        self.jungle = jungle
        self.kids = kids
        self.age = age

    def roar(self):
        print(f"{self.name} says WOOOAHHHH")

my_lion = Lion("simba", "savana", 4, 12)
my_lion.roar()

mufasa_lion = Lion("Mufasa", "Gir forest", 8, 18)
mufasa_lion.roar()
```

```python
## Modelling a bank account

# Define a class for bank account

class BankAccount:
    
    def __init__(self, accountholder, balance = 0):
        self.accountholder = accountholder
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} is deposited . New balance amount is {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"{amount} is withdrwan. New balance is {self.balance}")

    def get_balance(self):
        return self.balance
    

## Create an object

my_account = BankAccount("Deepak", 500000000)
print(my_account.balance)

your_account = BankAccount("Alizeh", 200000000000)
print(your_account.balance)

my_account.get_balance()



        
```
