```python
## BInary Files
## Writing to a binary file
## wb --> write byte mode
# # rb --> read byte mode 

bin_data = b'\x00\x01\x02\x03\x04\x05\x10'
with open('example2.bin', 'wb') as file:
    file.write(bin_data)
```

```python
## Check if path is a file or a dircetory

import os

path = 'test_dir'

if os.path.isfile(path):
    print(f"The path '{path}' is a file path")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory. ")
else:
    print(f"The path '{path}' is neither a file nor a directory")
```
