```python
## CSV  import csv
import csv
with open('example.csv', mode= 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Krish', 37])

with open('example.csv', mode = 'r' ) as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

FILE OPERATIONS - READ< WRITE>

```python
### Read a whole file
with open('source.txt', 'r') as file:
    for line in file:
        print(line)
```

```python
### Read a whole file
with open('source.txt', 'r') as file:
    for line in file:
        content = file.read()
        print(content)
```

```python
## Read the whole file

with open ('example.txt', 'r') as file:
    content=file.read()
    print(content)
```

```python
## Read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip()) ## .strip removes the new line spaces
```

```python
## Writing a File and (Overwriting)

with open('example_3.txt', 'w') as file:
    file.write('Hello World! New One\n')
    file.write('This is a new line \n')
```

```python
## Write a file without over writing
with open('example.txt', 'a') as file:
    file.write("Appending the file \n")
```

```python
## Writing a list of lines in a file
lines = ["First line \n", "Second Line \n", "Third Line \n"]

with open('example.txt', 'a') as file:
    file.writelines(lines)
```
