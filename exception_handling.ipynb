```python
## try , except, else block

try:
    num = int(input("Enter a number: "))
    result = 10 /num

except ValueError:
    print("This not a valid number")

except ZeroDivisionError:
    print("Enter denomniator greater than Zero.")

except Exception as ex:
    print(ex)
    print("Please talk to the admin ")
else:
    print(f"The result is {result}")
```

```python
## try , except, else, finally

try:
    num = int(input("Enter a number: "))
    result = 10 /num

except ValueError:
    print("This not a valid number")

except ZeroDivisionError:
    print("Enter denomniator greater than Zero.")

except Exception as ex:
    print(ex)
    print("Please talk to the admin ")
else:
    print(f"The result is {result}")
finally:
    print("Execution complete")


## Praatical usage of DB connection, error, excepttion and finally close the DB connection
```

```python
## Practical Examples

## File handling adn Exception Handling

try:
    file = open('example.txt', 'r')
    content = file.read()
    print(content)

except FileNotFoundError:
    print("The file is not yet created or does not exist.")

finally:
    if 'file' in locals() or not file.closed():
        print(file.close())
        print("File Closed") 
```
